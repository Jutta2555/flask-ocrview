```python
jsondatafile='/vsocr/GHB20/ok/2021/11/21/ocrdata_total.json'
with open(jsondatafile,'r',encoding='utf8') as f:
    j=json.load(f)

csv_dict_data=[]

for odic in j['ocrdata']:
    # print(json.dumps(odic,indent=4,ensure_ascii=False))
    # json.dumps(js,indent=4,ensure_ascii=False)
    # print(json.dumps(odic,indent=4,ensure_ascii=False))
    print(odic.keys())
    addressdict= {}
    for key in KEYSDICT:
        if key in odic:
            addressdict[key]=odic[key]
        else:
            addressdict[key]=''
    
    # print(addressdict)
    csv_dict_data.append(addressdict)    
```

```python

# 'เตรียมไฟล์ csv'
csv_file='/vsocr/GHB20/ok/2021/11/21/docrghb.csv'
# csv_columns=list(csv_dict_data[0].keys())
try:
    with open(csv_file, 'a') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=KEYSDICT)
        writer.writeheader()
        for data in csv_dict_data:
            writer.writerow(data)

except IOError:
    print("I/O error")
print('goto find zipcode ')
```
